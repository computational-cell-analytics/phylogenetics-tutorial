## Molecular Phylogenetics Tutorial

This exercise gives a short intodruction into molecular phylogenetics. Specicially, we will use a single gene to form a hypothesis for the phylogenetic tree of some primate species. We will use the cytochrome B gene (`MT-CYB / CYTB`), which is a well-established molecular marker for primate evolution. See [Farias et al.](https://link.springer.com/article/10.1007/s002390010197) for details.

We will use a hierarchical clustering approach ([UPMGA](https://en.wikipedia.org/wiki/UPGMA)) based on the normalized distance of the gene. This is a very simple approach, we will discuss limitations and more advanced techniques at the end.

```python
# The main imports for reading and processing the data.
import numpy as np
import pandas as pd
from Bio import SeqIO
```

```python
# Load the already prepared sequences for the MT-CYTB gene.
# The sequences for Human, Chimpanzee, Gorilla, Orangutan, Gibbon, and Macaque
# were downloaded from the NCBI GeneBoank already.
sequence_data_file = "data/primate_mt_cytb.fasta"
sequence_records = list(SeqIO.parse(sequence_data_file, format="fasta"))
```

```python
# Inspect the data.
for i, record in enumerate(sequence_records):
    print(f"Sequence {i}:", record.description, print(len(record.seq)))
```

```python
# Let's check an individual sequence.
sequence_records[0].seq
```

```python
# Compute the normalized hamming distance between two sequences:
# We define a function for this.
def compute_hamming_distance(seq1, seq2):
    # Prepare the inputs.
    seq1, seq2 = str(seq1).upper(), str(seq2).upper()
    # Make sure the sequences have the same length.
    assert len(seq1) == len(seq2)
    # Compute the normalized hamming distance.
    hamming_distance = 0.0
    # FILL IN: How do we compute the hamming distance? 
    # ...
    sequence_length = len(seq1)
    normalized_distance = hamming_distance / sequence_length
    return normalized_distance
```

```python
# Let's test it for one pair of sequences.
dist = compute_hamming_distance(sequence_records[0].seq, sequence_records[1].seq)
print(dist)
```

```python
# Let's compute the distance matrix.
# It contains the distance for each pair of species.
n_species = len(sequence_records)
species = [rec.id for rec in sequence_records]

distances = np.zeros((n_species, n_species))
for i in range(n_species):
    for j in range(i + 1, n_species):
        dist = compute_hamming_distance(
            sequence_records[i].seq, sequence_records[j].seq
        )
        # FILL IN: how do we write the distance for the given pair of species into the result matrix?
        # ...
distances = pd.DataFrame(distances, index=species, columns=species)
```

```python
# Inspect the distances
distances.round(4)
```

## Hierarchical clustering / UPMGA


Short intermission to explain the clustering approach.

```python
# Additional imports to implement and visualize the clustering result.
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
from scipy.spatial.distance import squareform
```

```python
# Create a condensed representation: 
# remove the main diagonal with self-distances
condensed_distances = squareform(distances.values)
# Run hierarchical clustering. 
# UPMGA corresponds to average clustering criterion.
clustering = linkage(condensed_distances)
```

```python
# Visualize the clustering result (= hypothesis for the phylogenetic tree).
plt.figure()
dendrogram(clustering, labels=species, orientation="bottom", link_color_func=lambda k: "black")
plt.ylabel("Distance")
plt.tight_layout()
plt.show()
```
